```python
from cyclegan.datamodules.apple2orange import Apple2OrangeDataModule
import matplotlib.pyplot as plt
```

## 4. Visualizing

```python
dm = Apple2OrangeDataModule('/content/data', num_workers=0)
```

### Train Set

```python
dm.setup()
dl_train = dm.train_dataloader()
for batch in dl_train:
  plt.imshow(batch[0][0].permute(1,2,0)); plt.show()
  plt.imshow(batch[1][0].permute(1,2,0)); plt.show()
  break
```

### Val Set

```python
dm.setup()
dl_val = dm.val_dataloader()
for batch in dl_val:
  plt.imshow(batch[0][0].permute(1,2,0)); plt.show()
  plt.imshow(batch[1][0].permute(1,2,0)); plt.show()
  break
```

### Test Set

```python
dm.setup('test')
dl_test = dm.test_dataloader()
for batch in dl_test:
  plt.imshow(batch[0][0].permute(1,2,0)); plt.show()
  plt.imshow(batch[1][0].permute(1,2,0)); plt.show()
  break
```
